```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

 
data = load_digits().data
pca = PCA(2)  #per visualizzare il plot del clustering 2D
 
df = pca.fit_transform(data)
 
df.shape
```

```python
from sklearn.cluster import KMeans
 
kmeans = KMeans(n_clusters= 10) #sappiamo che sono 10 cluster perch√® le label vanno da 0 a 9
 
label = kmeans.fit_predict(df) #effettua la clusterizzazione
 
print(label)
```

```python
filtered_zero = df[label==0]
print(filtered_zero.shape)

#plot del cluster con label 0
plt.scatter(filtered_zero[:,0] , filtered_zero[:,1])
plt.show()
```
